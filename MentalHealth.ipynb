```python
import matplotlib.pyplot as plt

# Extracting data for plotting
years = df['Year']
schizophrenia = df['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized']
depressive = df['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']
anxiety = df['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized']
bipolar = df['Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized']
eating = df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized']

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(years, schizophrenia, label='Schizophrenia', alpha=0.7)
plt.bar(years, depressive, label='Depressive', alpha=0.7)
plt.bar(years, anxiety, label='Anxiety', alpha=0.7)
plt.bar(years, bipolar, label='Bipolar', alpha=0.7)
plt.bar(years, eating, label='Eating', alpha=0.7)

plt.title('Trends of Mental Disorders Over Years')
plt.xlabel('Year')
plt.ylabel('Share of Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df containing your data

year = 2019
data_year = df[df['Year'] == year]

# Drop rows with NaN values
data_year = data_year.dropna()

# Extracting the values for each mental disorder
schizophrenia_value = data_year['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized'].sum()
depressive_value = data_year['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'].sum()
anxiety_value = data_year['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'].sum()
bipolar_value = data_year['Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized'].sum()
eating_value = data_year['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'].sum()

# Labels for each mental disorder
labels = ['Schizophrenia', 'Depressive', 'Anxiety', 'Bipolar', 'Eating']

# Values for each mental disorder
values = [schizophrenia_value, depressive_value, anxiety_value, bipolar_value, eating_value]

# Plotting the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(labels, values)
plt.title(f'Share of Mental Disorders in {year}')
plt.xlabel('Mental Disorder')
plt.ylabel('Share of Population')
plt.grid(True)
plt.show()
```
